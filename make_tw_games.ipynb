# tw-cooking games

Recipe(INT): Number of ingredients in the recipe.

Take(INT): Number of ingredients to find. (<= recipe)

Go(INT): Number of locations in the game. (1, 6, 9, or 12)

open/cook/cut/drop: T/F (Default: F)

recipe-seed: Random seed used for generating the recipe.(Default: 0)


```python
!tw-make tw-cooking --recipe 1 --take 1 --go 1 --open --cook --cut --drop --recipe-seed 1 --seed 1 --output tw-games/tw-cooking_Recipe-1_Take-1_Go-1.z8
```
